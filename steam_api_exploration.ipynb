### Steam data API exploration

```python
import requests
import pandas as pd
import os
```

```python
## variables

# Define the base URL for the Steam API
base_url = "https://api.steampowered.com/"

# Define your API key (replace 'YOUR_API_KEY' with your actual API key)
#api_key = "46DE8A18C6DAE9233CF6C24CFD93A4FF"
api_key = os.environ.get('steam_api_key')

# Define the Steam ID of the user you want to retrieve information for
#steam_id = "76561197997758075"
steam_id = os.environ.get('steam_api_id')

# Construct the full URL
# url = f"{base_url}{endpoint}?key={api_key}&steamids={steam_id}"
```

```python
def get_steam_data(url):
    # Send a GET request to the Steam API
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()
        return data
    else:
        print(response)
```

```python
general_data = get_steam_data(url = f"{base_url}ISteamUser/GetPlayerSummaries/v0002/?key={api_key}&steamids={steam_id}")
```

```python
store_games_list = get_steam_data(url = f"{base_url}ISteamApps/GetAppList/v0002/?key={api_key}&steamids={steam_id}")
store_games_list = pd.json_normalize(store_games_list['applist']['apps'])
```
